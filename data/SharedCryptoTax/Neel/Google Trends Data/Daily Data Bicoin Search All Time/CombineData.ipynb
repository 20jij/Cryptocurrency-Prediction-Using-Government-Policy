```python
import pandas as pd
```

### Combining gtrends Data

```python
ADtoAU = pd.read_csv("AD-AU.csv", nrows = 15627)
AWtoBH = pd.read_csv("AW-BH.csv", nrows = 14856)
BI = pd.read_csv("BI.csv")
BJtoBS = pd.read_csv("BJ-BS.csv")
```

```python
BJtoBS
```

```python
frames = [ADtoAU, AWtoBH, BI, BJtoBS]

result = pd.concat(frames)
```

```python
result
```

### Combining with Block Summary Data

```python
block_summary = pd.read_csv("block_summary_date.csv")
block_summary.describe()
```
