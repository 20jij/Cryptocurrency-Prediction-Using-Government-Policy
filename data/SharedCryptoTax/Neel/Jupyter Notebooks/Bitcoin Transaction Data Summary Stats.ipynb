```python
import pandas as pd
df = pd.read_excel('blockchair_bitcoin_transactions_20161001.xlsx')
```

```python
df
```

```python
counts = df.iloc[:,[0,2]].groupby('time').count()
counts = counts.rename(columns = {"block_id":"count"})
counts
```

```python
mean = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').mean().rename(columns = {"input_total_usd":"input_total_usd_mean", "output_total_usd":"output_total_usd_mean"})
median = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').median().rename(columns = {"input_total_usd":"input_total_usd_median", "output_total_usd":"output_total_usd_median"})
max = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').max().rename(columns = {"input_total_usd":"input_total_usd_max", "output_total_usd":"output_total_usd_max"})
min = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').min().rename(columns = {"input_total_usd":"input_total_usd_min", "output_total_usd":"output_total_usd_min"})
firstq = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').quantile(0.25).rename(columns = {"input_total_usd":"input_total_usd_firstq", "output_total_usd":"output_total_usd_firstq"})
thirdq = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').quantile(0.75).rename(columns = {"input_total_usd":"input_total_usd_thirdq", "output_total_usd":"output_total_usd_thirdq"})
sum = df.loc[:,["time", "input_total_usd", "output_total_usd"]].groupby('time').sum().rename(columns = {"input_total_usd":"input_total_usd_sum", "output_total_usd":"output_total_usd_sum"})
```

```python
mean
```

```python
median
```

```python
result = pd.concat([counts, min, max, mean, firstq, median, thirdq, sum], axis = 1)
result
```
